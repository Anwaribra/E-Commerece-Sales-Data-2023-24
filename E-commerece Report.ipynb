# E-commerce Sales Data 2023-24
# Introduction 
The E-commerce Sales Data 2023-24 dataset is a valuable resource for e-commerce businesses and data scientists seeking to improve the shopping experience, enhance customer satisfaction, and drive business growth through data-driven insights. 

The dataset contains information about product sales and user behavior in an e-commerce store over the period from 2023 to 2024. The core information in the dataset includes:

- **Product information**: Product name, price, description, category, and ratings.
- **User information**: Name, address, email, date, time, and user behavior.
- **Transaction information**: Purchase date, time, quantity, value, and payment type.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df[['Purchase Amount (USD)']]
y=df['Previous Purchases']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```
